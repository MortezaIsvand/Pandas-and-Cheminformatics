```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Chembl_library_2M_2.csv')
df.head()
```

## Get dropped rows

```python
df2 = df.dropna()      # DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
df2
```

```python
dropped_rows = df[~df.index.isin(df2.index)]
dropped_rows
```

```python
dropped_rows = df[np.invert(df.index.isin(df2.index))]
dropped_rows
```

```python
df3 = df.dropna(thresh=df.shape[1]*0.8, axis=0)
df3
```

```python
dropped_rows = df[~df.index.isin(df3.index)] # Get dropped rows by unary (invert) operation.
dropped_rows
```

```python
dropped_rows = df[np.invert(df.index.isin(df3.index))]
dropped_rows
```

```python
df4 = df.dropna(subset=['MolecularWeight'])
df4
```

```python
dropped_rows = df[~df.index.isin(df4.index)]
dropped_rows
```

# Get dropped columns

```python
df5 = df.dropna(thresh=len(df)*0.8, axis=1) 
df5
```

```python
dropped_columns = ~df.columns.isin(df5.columns)
dropped_columns
```

```python
dropped_columns = df.loc[:, ~df.columns.isin(df5.columns)]
dropped_columns
```
