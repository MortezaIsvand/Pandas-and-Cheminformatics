### Selecting rows and columns by labels and positions using loc[ ] and iloc[ ] functions(slicing)

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Chembl_library_2M.csv', sep=';')
df.head()
```

```python
df.loc[:, ['Synonyms', 'Name']]
```

```python
df.iloc[:, [2, 1]]
```

```python
df.loc[:, 'Name':'HBA'] # Select columns between two columns
```

```python
df.iloc[:, 1 : 11]
```

```python
df.loc[:, 'HBA':]        # Select columns by range
```

```python
df.iloc[:, 10:]
```

```python
df.loc[:, :'HBA']       # Select columns by range
```

```python
df.iloc[:, :11]
```

```python
df.iloc[:, -1]
```

```python
df.iloc[:, -1:]
```

```python
df.iloc[:, -3:]
```

```python
df.iloc[-1,]
```

```python
df.iloc[-1:]
```

```python
df.iloc[-3:]
```

```python
df.iloc[::4]
```

```python
df.describe()
```

```python
df.describe().loc['min':'max']
```
