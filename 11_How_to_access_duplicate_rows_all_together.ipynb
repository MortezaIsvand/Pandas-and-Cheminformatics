# How to access duplicate rows

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Chembl_library_2M_2.csv')
df.head()
```

```python
df2 = df[['ChEMBLID', 'MolecularWeight','AlogP', 'MolecularSpecies', 'Smiles']]
df2
```

```python
df2 = df[['ChEMBLID', 'Smiles', 'MolecularSpecies', 'MolecularWeight', 'AlogP' ]]
df2
```

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna(subset=['Smiles'])
df3
```

```python
dropped_rows = df2[~df2.index.isin(df3.index)]
dropped_rows
```

```python
df3.Smiles.value_counts()
```

```python
df3.duplicated(subset='Smiles')             # keep: {'first', 'last', False}, default 'first'
```

```python
df3.Smiles.duplicated()
```

```python
df3.Smiles.duplicated().sum()
```

```python
df3.duplicated(subset=['Smiles', 'MolecularSpecies']) 
```

## Method 1

```python
duplicated_smiles = df3[df3.duplicated(subset=['Smiles'])]
duplicated_smiles
```

```python
duplicated_smiles = df3[df3.duplicated(subset=['Smiles'], keep=False)]
duplicated_smiles
```

```python
duplicated_smiles = df3[df3.duplicated(subset=['Smiles'], keep=False)].sort_values('Smiles')
duplicated_smiles
```

## Method 2

```python
df4 = df3.drop_duplicates('Smiles')
df4
```

```python
dropped_rows = df3[~df3.index.isin(df4.index)]
dropped_rows
```

```python
duplicated_smiles = pd.concat([df4.loc[df4.Smiles.isin(dropped_rows.Smiles), :], dropped_rows]).sort_values('Smiles')
duplicated_smiles
```
